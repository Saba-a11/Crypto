```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of time steps
T = 1000

# initial proce
P0 = 10

# drift
mu = 0.001
```

```python
# last log price
last_p = np.log(P0)

log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
  # sample a log return
  r = 0.01 * np.random.randn()

  # compute the new log price
  p = last_p + mu + r

  # store the return and price
  log_returns[t] = r + mu
  prices[t] = np.exp(p)

  # assign last_p
  last_p = p
```

```python
plt.figure(figsize=(20, 8))
plt.plot(prices)
plt.title("Simulated Prices")
```

```python
# number of time steps
T = 1000

# initial proce
P0 = 10
C0 = 8.5
L0 = 12

# drift
muP = 0.001
muC = 0.0013
muL = 0.00098
```

```python
# last log prices
last_p = np.log(P0)
last_c = np.log(C0)
last_l = np.log(L0)

log_returns_p = np.zeros(T)
prices_p = np.zeros(T)

log_returns_c = np.zeros(T)
prices_c = np.zeros(T)

log_returns_l = np.zeros(T)
prices_l = np.zeros(T)

for t in range(T):
  # sample a log returns
  rP = 0.01 * np.random.randn()
  rC = 0.01 * np.random.randn()
  rL = 0.01 * np.random.randn()

  # compute the new log prices
  p = last_p + muP + rP
  c = last_c + muC + rC
  l = last_l + muL + rL

  # store the returns and prices
  log_returns_p[t] = rP + muP
  prices_p[t] = np.exp(p)

  log_returns_c[t] = rC + muC
  prices_c[t] = np.exp(c)

  log_returns_l[t] = rL + muL
  prices_l[t] = np.exp(l)

  # assign last price
  last_p = p
  last_c = c
  last_l = l
```

```python
plt.figure(figsize=(20, 8))
plt.plot(prices_p)
plt.plot(prices_c)
plt.plot(prices_l)
```
