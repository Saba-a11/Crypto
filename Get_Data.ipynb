```python
curl = "https://poloniex.com/public?command=returnChartData&currencyPair=BTC_XMR&start=1546300800&end=1546646400&period=14400"
```

```python
import requests
import json
import pandas as pd

req = requests.get(curl)
data = json.loads(req.text)
df = pd.DataFrame(data)
df
```
